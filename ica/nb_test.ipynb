```python
import numpy as np
import numpy.fft as nfft
from scipy.fft import fft, ifft, rfft, irfft
from scipy.signal import windows
import matplotlib.pyplot as plt

```

```python
size = 100
x = np.arange(size)
ham = windows.get_window('hamming', size, True)

plt.plot(x, ham)
```

```python
iham = nfft.ifftshift(ham)
iham
plt.plot(x, iham)
```

```python
hamfft = nfft.rfft(ham)
hamfft = nfft.ifftshift(hamfft)
k = nfft.rfftfreq(size)*size
# print(hamfft)

plt.plot(k, hamfft)
```

```python
g = np.array([1.0, 2.0, 1.0, -1.0, 0.0, -2.0, 1.5, -2.5, 1.0, 0.5, -1.0, -1.5])
x = nfft.rfft(g)
N = g.size

k_low, k_up = 0.0, N//2+1
k_low = np.ones(np.shape(x))*k_low
k_up = np.ones(np.shape(x))*k_up

k = np.fft.rfftfreq(N)*N

print(g.size)
print(x)
print(k)
print(k_low)
print(k[:N//2+1])
print(np.less_equal(k, k_up))
print(np.greater_equal(k, k_low))
print(x)
x = np.where(np.logical_and(np.less_equal(k, k_up), np.greater(k, k_low)), x, 0.0)
print(x)
x_inv = np.fft.irfft(x)

# print(x)
# print(x_inv)
```
