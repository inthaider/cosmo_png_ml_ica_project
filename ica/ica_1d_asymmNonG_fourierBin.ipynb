## Setup for dividing zeta(G) + zeta(NG) into two (in Fourier space), and treating each half as a separate observed sample

```python
# Note that 'g' is the field used to generate the NG component called 'NG_asymsinh', 
# so 'NG_asymsinh' is correlated with 'g', and 'f' is a different Gaussian random 
# field with which the 'NG_asymsinh' is uncorrelated. 

source_g = f
source_ng = NG_asymsinh
size_field = source_g.size
correlated = False

source = source_g + source_ng

source_g1 = grf.window(source_g, N, kc[0], kc[2])
source_g2 = grf.window(source_g, N, kc[2], kc[4])
source_g3 = grf.window(source_g, N, kc[4], kc[6])

source_ng1 = grf.window(source_ng, N, kc[0], kc[3])
source_ng2 = grf.window(source_ng, N, kc[3], kc[6])
source_ng3 = grf.window(source_ng, N, kc[6], kc[9])

source1 = grf.window(source, N, kc[0], kc[3])
source2 = grf.window(source, N, kc[3], kc[6])
source3 = grf.window(source, N, kc[6], kc[9])

num_comps = 2
num_samples = 3

mix_signal_pre = np.vstack([source1, source2, source3])
```
